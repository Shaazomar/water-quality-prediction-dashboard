### When building an ML model, class balance is key:
    •If one class dominates (like Unsuitable here), the model may get lazy and always predict that class.
	•Example: If 70% of wells are “Unsuitable,” a dumb model that always predicts “Unsuitable” will still get 70% accuracy but it’s useless.


```python
X
```

### Spliting the data

## Intuition Behind This Step
	•If we split randomly without stratify → “Excellent” (only ~4% of data) might vanish from the test set.Then our model would never be tested on it → giving us false confidence.
	•Stratification ensures fair evaluation: the test set is a “miniature version” of the full dataset.


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

## Training the model

```python
# ===========================
# Logistic Regression with Class Weights
# ===========================

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Step 1: Compute class weights based on training data
classes = np.unique(y_train)  # unique class labels in y_train
class_weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)

# Convert to dictionary format {class_label: weight}
class_weights = dict(zip(classes, class_weights))
print("Class Weights:", class_weights)

# Step 2: Train Logistic Regression with class weights
log_reg = LogisticRegression(class_weight=class_weights, max_iter=500, random_state=42)
log_reg.fit(X_train, y_train)

# Step 3: Predict on the test set
y_pred = log_reg.predict(X_test)

# Step 4: Evaluate performance
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
