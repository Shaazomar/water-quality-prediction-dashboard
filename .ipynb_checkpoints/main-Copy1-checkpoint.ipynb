## Stratified shuffle split for spliting data into train and test set

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_idx, test_idx in split.split(df, df['Potability']):
    train_set = df.loc[train_idx]
    test_set = df.loc[test_idx]
```

## Seperating labels and feautures

```python
X_train = train_set.drop("Potability", axis=1)
y_train = train_set["Potability"].copy()
```

```python
X_test = test_set.drop("Potability", axis=1)
y_test = test_set["Potability"].copy()
```

### Filling Missing values and Pipeline

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])
```

```python
num_pipeline
```

```python
water_prepared = num_pipeline.fit_transform(X_train)   # fit + transform train
```

```python
water_test_prepared = num_pipeline.transform(X_test)
```

## Testing Different ML Algorithms

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

dec_cl = DecisionTreeClassifier(random_state=42)
rand_cl = RandomForestClassifier(random_state=42)

dec_cl.fit(water_prepared, y_train)
rand_cl.fit(water_prepared, y_train)

```

```python
dec_cl.fit(water_prepared, y_train)
```

## Evaluate on Test Set

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predictions on test set
dec_preds = dec_cl.predict(water_test_prepared)
rand_preds = rand_cl.predict(water_test_prepared)

# Evaluation
print("Decision Tree Accuracy:", accuracy_score(y_test, dec_preds))
print("Decision Tree Precision:", precision_score(y_test, dec_preds))
print("Decision Tree Recall:", recall_score(y_test, dec_preds))
print("Decision Tree F1 Score:", f1_score(y_test, dec_preds))

print("Random Forest Accuracy:", accuracy_score(y_test, rand_preds))
print("Random Forest Precision:", precision_score(y_test, rand_preds))
print("Random Forest Recall:", recall_score(y_test, rand_preds))
print("Random Forest F1 Score:", f1_score(y_test, rand_preds))
```

### Add Cross-Validation for your current models

```python
from sklearn.model_selection import cross_val_score

# Cross-validation for Decision Tree
dt_scores = cross_val_score(dec_cl, water_prepared, y_train, cv=5, scoring='accuracy')
print("Decision Tree CV Scores:", dt_scores)
print("Decision Tree Mean Accuracy:", dt_scores.mean())

# Cross-validation for Random Forest
rf_scores = cross_val_score(rand_cl, water_prepared, y_train, cv=5, scoring='accuracy')
print("Random Forest CV Scores:", rf_scores)
print("Random Forest Mean Accuracy:", rf_scores.mean())
```
