## Step 1: Import the Basic Libraries

```python
# Basic libraries for data handling
import pandas as pd
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Your Dataset

```python
df = pd.read_csv("waterDataset.csv")  
df.head()
```

## Step 3: Basic Info

```python
# Check basic info
df.info()

# Quick statistics
df.describe()
```

## Step 4: Missing Values Check

```python
df.isnull().sum()
```

## Step 5: Class Balance (Target Distribution)

```python
df['Water Quality Classification'].value_counts()
```

## Step 6: Quick Visualization of Classes

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Water Quality Classification', order=df['Water Quality Classification'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Distribution of Water Quality Classes")
plt.show()
```

## Step 7 Quick feature statistics & distributions

```python
df.hist(figsize=(15,10), bins=30)
plt.show()
```

## Step8:Correlation heatmap (important for seeing redundant features)

```python
plt.figure(figsize=(10,6))

# Select only numeric columns for correlation
corr_matrix = df.select_dtypes(include=['number']).corr()

sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
